```python
import pandas as pd
df=pd.read_csv('covid_toy.csv')
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('has_covid',axis=1),df['has_covid'],test_size=0.2,random_state=0)
X_train.shape,X_test.shape
```

# Manual Preprocessing of data

### Treating null values

```python

df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
si.fit(X_train[['fever']])
X_train_fever=si.transform(X_train[['fever']])
X_test_fever=si.transform(X_test[['fever']])
```

### Treating Categorical values

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Mild','Strong']])
oe.fit(X_train[['cough']])
X_train_cough=oe.transform(X_train[['cough']])
X_test_cough=oe.transform(X_test[['cough']])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
ohe.fit(X_train[['gender','city']])
X_train_gender_city=ohe.transform(X_train[['gender','city']])
X_test_gender_city=ohe.transform(X_test[['gender','city']])

```

### Extracting age

```python
X_train_age=X_train.drop(columns=['gender','fever','cough','city'])
X_test_age=X_test.drop(columns=['gender','fever','cough','city'])
```

### Concatenating data columnns

```python
import numpy as np
X_train_trf=np.concatenate((X_train_age,X_train_gender_city,X_train_cough,X_train_fever),axis=1)
X_test_trf=np.concatenate((X_test_age,X_test_gender_city,X_test_cough,X_test_fever),axis=1)
X_train_trf=pd.DataFrame(X_train_trf)
X_train_trf
```
